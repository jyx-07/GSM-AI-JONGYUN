```python
from random import *

# 난수 생성
data_1 = []
data_2 = []
for i in range(0, 10):
    data_1.append(uniform(0,1))
for i in range(0, 10):
    data_2.append(uniform(0,2))
    
# DataFrame 생성
dict_data = {'first': data_1,
             'second': data_2}
index_data = ['zero', '1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th']
df = pd.DataFrame(dict_data, index = index_data)
df
```

```python
# 전체 평균
mean_val = df.mean()
# first 평균
first_mean = df['first'].mean()
# second 평균
second_mean = df['second'].mean()

print(mean_val)
print('\n')
print(f'first 평균: {first_mean}')
print(f'second 평균: {second_mean}')
```

```python
# 전체 중간값
median_val = df.median()
# first 중간값
first_median = df['first'].median()
# second 중간값
second_median = df['second'].median()

print(median_val)
print('\n')
print(f'first 중간값: {first_median}')
print(f'second 중간값: {second_median}')
```

```python
# 전체 최대값
max_val = df.max()
# first 최대값
first_max = df['first'].max()
# second 최대값
second_max = df['second'].max()

print(max_val)
print('\n')
print(f'first 최대값: {first_max}')
print(f'second 최대값: {second_max}')
```

```python
# 전체 최소값
min_val = df.min()
# first 최소값
first_min = df['first'].min()
# second 최소값
second_min = df['second'].min()

print(min_val)
print('\n')
print(f'first 최소값: {first_min}')
print(f'second 최소값: {second_min}')
```

```python
# 전체 표준편차
std_val = df.std()
# first 표준편차
first_std = df['first'].std()
# second 표준편차
second_std = df['second'].std()

print(std_val)
print('\n')
print(f'first 표준편차: {first_std}')
print(f'second 표준편차: {second_std}')
```

```python
def plot(fs, dpi):
    fig, ax = plt.subplots(figsize=fs, dpi=dpi)
    ax.set_title(f"사이즈: {fs}, dpi: {dpi}")
    ax.plot([0, 1, 2], [4, 10, 6])
    
for i in range(1, 4):
    plot((2, 2), i*72)
    
for i in [2, 4, 6]:
    plot((i, i), 72)
```
